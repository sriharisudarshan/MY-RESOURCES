```python
car_sales_missing = pd.read_csv("car_sales_missing.csv")
```

```python
car_sales_missing
```

```python
# Fills missing fields under Odometer with mean of present Odometer values
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
car_sales_missing
```

```python
# Inplace parameter eliminates the need for reasignment and allows
# changes to be made to the original data frame without reassignment as 
# done in [43]
car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(), inplace = True)
```

```python
car_sales_missing
```

```python
car_sales_missing.dropna()
```

```python
# Drops rows with missing rows
car_sales_missing.dropna(inplace = True)
```

```python
# Drops rows with missing rows
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped
```

```python
car_sales_missing_dropped.to_csv("car_sales_missing_dropped.csv")
```

```python
# Column from series
seats_column = pd.Series([5,5,5,5,5])

# New column called seats
car_sales["Seats"] = seats_column
```

```python
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace = True)
```

```python
car_sales
```

```python
# Column from python lists
# When creating columns from lists, the lenght of the list should be
# same as the number of rows in the data frame.
# This is not required when creating columns from Series as done in 
# the previous steps
fuel_economy = [7.5, 9.3, 5.0, 9.6, 8.7, 4.7, 7.6, 8.7,3.0,4.5]
car_sales["Fuel Economy per 100km"] = fuel_economy
car_sales
```

```python
# Fuel Use = Odometer reading * Mileage
car_sales["Total Fuel Use"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel Economy per 100km"]
```

```python
car_sales
```

```python
# Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
# New column with bool data type
car_sales["Passed road safety"] = True
```

```python
car_sales
```

```python
# Removing/Dropping a column
car_sales = car_sales.drop("Total Fuel Use", axis =1)
```
