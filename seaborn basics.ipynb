## seaborn

```python
import seaborn as sns
```

```python
a=sns.load_dataset("flights")
a.head()
```

#### Scatter plot

```python
sns.relplot('year','passengers',data=a) #replot is a function to display scatter-plot. year=x-axis, passenger=y-axis
```

#### Bringing third catogory to the chart and color-coding

```python
sns.relplot('year','passengers',hue='month',data=a)# hue=z, where z is a 3rd variable, this brings more clarity to the graph
```

#### To have different size for plotting

```python
sns.relplot('year','month',size='passengers',data=a) #size command brings the variation in the size of the plot
```

#### To change the size from a given range

```python
sns.relplot('year','month',size='passengers',sizes=(25,200),data=a)# use size command to mention the range, here 25-200 is range
```

```python
b=sns.load_dataset("titanic")
b.head()
```

#### Line chart

```python
sns.relplot('deck','fare',data=b,kind='line') #kind=line command, plots the graph as a line graph
```

#### To plot actual values without the mean

```python
sns.relplot('deck','fare',data=b,kind='line',estimator=None)
```

#### To print the graph as sum of values

```python
sns.relplot('deck','fare',data=b,kind='line',estimator=sum)
```

#### To involve one more parameter to the graph

```python
sns.relplot('deck','fare',data=b,kind='line',hue='sex',ci=None)# My 3rd parameter is added using hue, 
#ci is used to remove confidance interval
```

#### To put different graphs baased on one parameter

```python
sns.relplot('deck','fare',data=b,kind='line',hue='sex',col='embark_town',estimator=None) # different plots are displayed 
#according to different types of embark_town
```

#### To display only 2 graphs in a row

```python
sns.relplot('deck','fare',data=b,kind='line',hue='sex',col='embark_town',estimator=None,col_wrap=2)
#col_wrap is the function which helps to put desired number of graphs in particular row
```

```python
c=sns.load_dataset("car_crashes")
c.head()
```

#### catplot or swarm plot

```python
sns.catplot('abbrev','alcohol',data=c)
```

#### To not to show magnitute

```python
sns.catplot('abbrev','alcohol',data=c,jitter=False) #jitter is used to show the magnitude.
# note: there is no much difference in this dataset
```

#### To plot the graph for only a given condition

```python
sns.catplot('abbrev','alcohol',data=c.query("ins_premium > 1000"),jitter=False)#dataset.query('condition') helps to plot the 
#graph on;y for the given condition
```

#### To add swarm feature

```python
sns.catplot('abbrev','ins_losses',data=c.query('alcohol > 5'),kind='swarm') # to show the density
```

```python
d=sns.load_dataset("planets")
d.head()
```

#### box plot

```python
sns.catplot('year','method',data=d,kind='box') # to print in box format
```

#### To add 3rd parameter to the graph

```python
sns.catplot('year','method',data=d,hue='number',kind='swarm')# same graph is displayed in swarm for clarity purpose
```

```python
sns.catplot('year','method',data=d,hue='number',kind='box')#box display of above swarm graph
```

#### To display the graph based on the particular hue

```python
sns.catplot('year','method',data=d,hue='number',hue_order=[1,3,5,7],kind='box')#hue_order used to display only particual datas
```

#### plot different graphs based on a condition and in rows

```python
sns.catplot('year','method',data=d.query('mass > 20'),hue='number',kind='swarm',row='mass')
```

#### plot different graphs based on a condition and in columns

```python
sns.catplot('year','method',data=d.query('mass > 18'),hue='number',kind='swarm',col='mass',col_wrap=4)
```

#### Violin plot

```python
e=sns.load_dataset("iris")
e.head()
```

```python
sns.catplot('species','petal_length',data=e,kind='violin')
```

#### To identitify the density inside the plot

```python
sns.catplot('species','petal_length',data=e,kind='violin',inner='stick')# shows the density of the graph.
```

#### bar chart

```python
sns.catplot('sepal_length','petal_length',data=e,kind='bar')
```

#### highlight the chart based on one catogory

```python
sns.catplot('petal_width','petal_length',data=e,kind='bar',col='species',hue='sepal_width',col_wrap=2)
```

```python
sns.catplot('sepal_width','sepal_length',data=e,kind='bar',row='species')
```

#### Point plot

```python
sns.catplot('petal_length','species',data=e,kind='point',hue='sepal_width',sizes=(20,4000))
```

#### distribution plot

```python
sns.distplot(e.sepal_width)# take only one parameter and plots the graph
```

#### to remove the curve

```python
sns.distplot(e.sepal_width,kde=False)
```

#### To remove the bar

```python
sns.distplot(e.sepal_width,hist=False)
```

#### To show number of data points

```python
sns.distplot(e.sepal_length,kde=False,rug=True)
```

#### To adjust the graph scale

```python
sns.distplot(e.sepal_length,kde=False,rug=True,bins=50)
```

#### To plot graph based on condition

```python
sns.distplot(e.sepal_length[e.species=="setosa"],kde=False,rug=True)
```

#### Joint plot

```python
sns.jointplot('sepal_width','sepal_length',data=e)
```

#### using hex function

```python
sns.jointplot('sepal_width','sepal_length',data=e,kind='hex')
```

#### using kde 

```python
sns.jointplot('petal_width','sepal_length',data=e,kind='kde',color='#77ffff')
```

#### Pair plot

```python
sns.pairplot(e)
```

#### To add condition

```python
sns.pairplot(e,hue='species')
```
